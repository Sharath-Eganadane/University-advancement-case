Decision Tree Classification

```python
Data_Req = MasterData1.copy()
```

```python
#Creating dataframe using specific columns required for training

Data_Req = Data_Req[['Gender','Relationship Types','Faculty of Graduation','Province','Lifetime Giving','Year of Graduation','Events Attended in Lifetime','# of Instances of Volunteering','# of Clicks in the Past Month']]
Data_Req.head()
```

```python
Data_Req.dropna(inplace=True)
```

```python
Data_Req = Data_Req.reset_index(drop=True)
print(Data_Req.shape)
```

```python
#Encoding on the categorical features of the data

ordinal_encoding_features=['Events Attended in Lifetime','# of Instances of Volunteering','Year of Graduation','# of Clicks in the Past Month']
one_hot_encoding_features=['Gender','Relationship Types','Faculty of Graduation','Province']
```

```python
#Splitting data into
#X - Features
#Y - Target variable

y=Data_Req['Lifetime Giving']
Data_Req.drop('Lifetime Giving',axis=1,inplace=True)
```

```python
import sklearn
from sklearn import preprocessing
```

```python
#Creating encoder object

label_encoder = sklearn.preprocessing.LabelEncoder()

#Transform the data by fitting encoder object

y_encoded = label_encoder.fit_transform(y)

print(y_encoded)
```

```python
label_encoder.classes_
```

```python
#Creating encoder object
ordinal_encoder = sklearn.preprocessing.OrdinalEncoder()

#Transform the data by fitting encoder object

Data_Req[ordinal_encoding_features] = ordinal_encoder.fit_transform(Data_Req[ordinal_encoding_features])
```

```python
#Creating encoder object

one_hot_encoder = sklearn.preprocessing.OneHotEncoder()

# fit the encoder object and transform the data, save as encoded_columns

encoded_columns = one_hot_encoder.fit_transform(Data_Req[one_hot_encoding_features]).toarray()

## create names for the newly encoded features (this will use our original names + underscore + the binary category)

New_Col_Names = one_hot_encoder.get_feature_names_out(one_hot_encoding_features)

# print out the feature_names so you understand what they represent 

print(New_Col_Names)
```

```python
#Creating encoder object

Hot_Encoded_Columns_df = pd.DataFrame(encoded_columns, columns = New_Col_Names)

## drop the old columns and append the new columns

Partially_Encoded_Data = Data_Req.drop(one_hot_encoding_features, axis = 1)

Completely_Encoded_Data = pd.concat([Partially_Encoded_Data.reset_index(drop=True), Hot_Encoded_Columns_df.reset_index(drop=True)], axis = 1)
```

```python
print(Data_Req.shape)
print(Completely_Encoded_Data.shape)
```

```python
Data_Req[one_hot_encoding_features]
```

```python
#Splitting Data into Train and Test Data

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(Completely_Encoded_Data, y_encoded, test_size = 0.3, train_size = 0.7)
```

```python
from statistics import mode
# get the mode of the target variable
baseline = mode(y_train)
print('Mode of Target Variable:', baseline)
## create a vector of length y_test where each entry is the mode of the target variable
baseline_pred = baseline * np.ones(len(y_test))
## print the accuracy of this baseline model 
print(round(sklearn.metrics.accuracy_score(baseline_pred, y_test)*100, 2))
```

```python
from sklearn import tree
Decision_Tree_model = tree.DecisionTreeClassifier()
Decision_Tree_model.fit(X_train, y_train)
```

```python
print(Decision_Tree_model.feature_importances_)
print(X_train.columns)
```

```python
print('Tree depth: ', Decision_Tree_model.get_depth())
print('Number of leaf nodes: ', Decision_Tree_model.get_n_leaves())
```

```python
#Checking Training set accuracy

print('Training Set Accuracy: ', round(Decision_Tree_model.score(X_train, y_train) * 100, 2))
```

```python
Decision_Tree_model_predictions = Decision_Tree_model.predict(X_test)
print(Decision_Tree_model_predictions)
print(y_test)
```

```python
print('Test Set Accuracy: ', round(Decision_Tree_model.score(X_test, y_test)*100, 2))
```

```python
max_depth = 7
Decision_Tree_model_params = tree.DecisionTreeClassifier(max_depth = max_depth)
Decision_Tree_model_params.fit(X_train, y_train)
 
print('Tree depth: ', Decision_Tree_model_params.get_depth())

print('Number of leaf nodes: ', Decision_Tree_model_params.get_n_leaves())

print('training set accuracy: ', round(Decision_Tree_model_params.score(X_train, y_train) * 100, 2))

print('test set accuracy: ', round(Decision_Tree_model_params.score(X_test, y_test)*100, 2))
```

Training Data with Gini Criterion Parameter

```python
max_depth = 7
criterion = 'gini'
Decision_Tree_model_params_gini = tree.DecisionTreeClassifier(max_depth = max_depth, criterion = criterion)
Decision_Tree_model_params_gini.fit(X_train, y_train)
 
print('Tree depth: ', Decision_Tree_model_params_gini.get_depth())

print('Number of leaf nodes: ', Decision_Tree_model_params_gini.get_n_leaves())

print('training set accuracy: ', round(Decision_Tree_model_params_gini.score(X_train, y_train) * 100, 2))

print('test set accuracy: ', round(Decision_Tree_model_params_gini.score(X_test, y_test)*100, 2))
```

```python
plt.figure(figsize=(30,30))
tree_plot = tree.plot_tree(Decision_Tree_model_params_gini, fontsize = 8, feature_names = X_train.columns, filled = True, max_depth = 7)
plt.savefig("Tree_Plot.pdf", format="pdf", bbox_inches="tight")
plt.show()
```
